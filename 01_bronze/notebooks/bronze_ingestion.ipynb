# Bronze Layer - Data Ingestion

This notebook ingests raw data from AWS S3 into the Bronze layer using Delta tables.
No transformations are applied in this layer as per Medallion Architecture principles.

## Set Catalog and Schema

```python
%sql
use catalog 'e_commerce';
use schema bronze ;
```

Load Customers Table

```python
customers_df=spark.read.format("csv").option("header","true").option("inferSchema","false").load("s3://olist-de-miniproj-bucket/raw/olist_customers_dataset.csv")
display(customers_df)
customers_df.write.format("delta").mode("overwrite").saveAsTable("e_commerce.bronze.bronze_customers")
```

Load Order_reviews Table

```python
import pandas as pd
import io
 
s3_path = "s3://olist-de-miniproj-bucket/raw/olist_order_reviews_dataset.xlsx"
 
file_content = spark.read.format("binaryFile").load(s3_path).collect()[0]["content"]
 
pandas_df = pd.read_excel(io.BytesIO(file_content))
 
for col in pandas_df.columns:
    if pandas_df[col].dtype == 'object':
        pandas_df[col] = pandas_df[col].astype(str)
 
order_reviews_df = spark.createDataFrame(pandas_df)
 
order_reviews_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true")\
    .saveAsTable("e_commerce.bronze.bronze_order_reviews")

display(order_reviews_df)
```
