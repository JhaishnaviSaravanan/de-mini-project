```python
import pandas as pd
import io
 
s3_path = "s3://olist-de-miniproj-bucket/raw/olist_order_reviews_dataset.xlsx"
 
file_content = spark.read.format("binaryFile").load(s3_path).collect()[0]["content"]
 
pandas_df = pd.read_excel(io.BytesIO(file_content))
 
for col in pandas_df.columns:
    if pandas_df[col].dtype == 'object':
        pandas_df[col] = pandas_df[col].astype(str)
 
order_reviews_df = spark.createDataFrame(pandas_df)
 
order_reviews_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true")\
    .saveAsTable("e_commerce.bronze.bronze_order_reviews")

display(order_reviews_df)

 # order_reviews_df = spark.read.format("com.crealytics.spark.excel") \
#     .option("header", "true") \
#     .option("inferSchema", "false") \
#     .option("dataAddress", "'Sheet1'!A1") \
#     .load("s3://your-bucket-path/olist_order_reviews_dataset.xlsx")
# order_reviews_df.write.format("delta") \
#     .mode("overwrite") \
#     .saveAsTable("e_commerce.bronze.bronze_order_reviews")

# order_reviews_df=spark.read.format("csv").option("header","true").option("inferSchema","true").load("s3://olist-de-miniproj-bucket/raw/olist_order_reviews_dataset.csv")
# display(order_reviews_df)
# order_reviews_df.write.format("delta").mode("overwrite").saveAsTable("e_commerce.bronze.bronze_order_reviews")
```
